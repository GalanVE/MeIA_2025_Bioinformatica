# Aprendizaje supervisado

**El aprendizaje supervisado** ocurre cuando nuestro modelo se entrena en un conjunto de datos preetiquetados y desarrolla sistemas para predecir resultados de datos nuevos.

Los datos utilizados para entrenar el sistema en aprendizaje supervisado contienen valores de entrada y salida.

Este tipo de aprendizaje automático se clasifica normalmente en clasificación y regresión.

En la clasificación, la salida tiene valores discretos. Nuestro sistema tendrá que clasificar estos valores en sus respectivos grupos correctamente. Un ejemplo es clasificar los SNP en clase funcional o benigna, o predecir que el día será caluroso o frío.

En la regresión, la salida contiene valores continuos, por lo que el sistema tiene que predecir un valor de salida cercano al valor real. Un ejemplo es la predicción de la temperatura, los precios de las acciones y más.

### Evaluación de los modelos

- Un paso fundamentan en el desarrollo de algoritmos de machine learning, es la evaluación del modelo.

- La mejor manera es no utilizar todos los datos durante el entrenamiento. Por el contrio se aconseja dividir el conjunto de completo de datos en dos partes: el 80% de la información se utiliza para entrenar el algoritmo y el 20% de los datos se usa para probar que tan correctamente puede predecir el modelo.

- Este es el paso donde se determina la eficiencia de nuestro sistema.

- Para comprobar la precisión del modelo, comparamos los valores predichos con los valores reales en una matriz de confusión.

![texto del vínculo](https://images.datacamp.com/image/upload/v1701364260/image_5baaeac4c0.png)


A partir de la matriz de confusión podemos calcular diferentes métricas, que nos permiten evaluar el modelo.

- Accuracy
  - La exactitud es la tasa de predicciones correctas para un modelo o el número de valores predichos correctamente dividido por el número total de instancias en el conjunto de prueba:
  $Accuracy=\frac{TP+TN}{TP+FP+TN+FN}$

- Precision
 - La precisión es la proporción de verdaderos positivos y el número total de casos predichos como positivos por el modelo
  $P = \frac{TP}{TP + FP}$

- Recall
  - El recall es la tasa de verdaderos positivos y también se denomina sensibilidad de un modelo o el número de verdaderos positivos dividido por el número total de casos positivos en el conjunto de prueba.

  ![texto del vínculo](https://www.researchgate.net/publication/346129022/figure/fig3/AS:961239823970323@1606188841513/Visualizing-accuracy-recall-aka-sensitivity-and-precision-which-are-the-common_W640.jpg)

- F1 score
  - Es la medida armonica de la precision y el recall.
  $F1Score = 2*((Precision*Recall)/(Precision + Recall))$

### Receiver Operating Characteristics (ROC) Curve

La curva de característica operativa del receptor, es otro método para evaluación de los clasificadores. Es una gráfica entre la tasa de verdades positivos TPR o sensibilidad y la tasa de falsos positivos (FPR)(1- especificidad).

$TPR=\frac{TP}{TP+FN}$

$FPR=\frac{FP}{FP+TN}$


![texto del vínculo](https://upload.wikimedia.org/wikipedia/commons/thumb/1/13/Roc_curve.svg/1024px-Roc_curve.svg.png)

La curva sólida es el cambio de la tasa de verdaderos positivos con la tasa de falsos positivos. Cuento mayor sea el área bajo la curva, mayor será la accuracy del modelo.

La métricas de evaluación de los modelos de regresión son bastante complejas. La regresión trata con un conjunto de datos continuo, lo que significa que se requieren métricas avanzadas para su evaluación. Las métricas comunes para la evaluación del modelo de regresión son la varianza, específicamente, el errir cuadrático medio y el coeficiente de R cuadrado.

### Validación cruzada
Al realizar cualquier tarea de aprendizaje automático, normalmente, dividimos nuestro conjuntno de datos al azar en dos conjuntos (80-20), es decir conjunto de entrenamiento y de prueba.

Sin embargo, esta técnica tiene una limitación, si sacrificamos el 20% de nuestros datos para evaluar el sistema, es posibe que nuestro modelo no pueda entrenarse adecuadamente para los desafíos del mundo real. Es posible perder información valiosa cuando dividimos el conjunto de datos, este problema empeora cuando tenemos dataset pequeños.

Para evitar esto, utilizamos una técnica llamada validación cruzada. En esta técnica, dividimos y evaluamos nuesto modelo con varias combinaciones de estos subconjuntos. Estos metodos incluyen k-fold validation y Leave-One-Out Cross-validation, entre otros.

![texto del vínculo](https://www.sharpsightlabs.com/wp-content/uploads/2024/02/5-fold-cross-validation_SIMPLE-EXAMPLE_v2.png)

## Optimización de modelos
Es un paso crucial para mejorar la precisión de los resultados previos. Los modelos de machine learning se entrenan em conjuntos de datos completos; por lo tanto, tienen varios hiperparámetros. La selección del modelo y los valores de los parámetros correctos son hallazgos clave al ajustar un modelo.

Generalmente, los modelos se entrenan y evalúan de forma iterativa utilizando varias combinaciones de parámetros. Al mejorar nuestros modelos con regularidad y probar la precisión del modelo de forma continua con varias combinaciones de hiperparámetros, podemos garantizar que el rendimiento se pueda optimizar aún más.

### Busqueda de cuadricula (Grid search)
Es un método para escanear el espacio de todas las combinaciones de hiperparámetros de un modelo para descubrir la mejor combinación de hiperparámetros. Dependiendo del tipo de modelo utilizadao, diferents hiperpárametros son necesarios.

La búsqueda de cuadrícula se puede aplicar a cualquier modelo de aprendizaje automático para calcular los mejores hiperparámetros para usar en ese sistema. Es importante señalar que la búsqueda en cuadrícula requiere un uso intensivo de computo y puede llevar mucho tiempo ejecutarla en un modelo. Iterar a través de cada combinación de hiperparámetros posible y almacena un modelo para cada combinación.

### Búsqueda aleatoria

James Bergstra y Yoshua Bengio (2012) propusieron la idea de búsqueda aleatoria de hiperparámetros en un sistema. Este tipo de búsqueda es completamente diferente del enfoque de cuadrícula. En lugar de examinar todas las combinaciones posibles en el espacio de hiperparámetros, una búsqueda aleatoria solo selecciona unos pocos puntos de muestra de la distribución y realiza los cálculos en esos puntos.

![texto del vínculo](https://static.wixstatic.com/media/fd32e3_50364e42770b42c28e3d9837487e12d1~mv2.png)

### Métodos de Ensemble
Se trata de una combinación de procedimientos en los que se utilizan varios modelos juntos para formar una versión significativamente mejorada. Esto nos permite adquirir mejores resultados en comparación con los modelos individuales.

Los dos métodos de conjuntos más utilizados son el promediado y la votación. Estos no son complejos y pueden implementarse fácilmente para aumentar la precisión del sistema.

El promedio se realiza cuando se trata de rangos de datos continuos, mientras que la votación se realiza cuando se trata de una clasificación discreta de datos. Inicialmente, se entrenan varios modelos diferentes mediante diferentes subconjuntos de datos de entrenamiento. A continuación, se combinan sus resultados. Por lo tanto, en lugar de crear un solo sistema y luego asumir que es el más preciso, los métodos de conjunto adoptan varios sistemas y los promedian para producir un producto final.

## Desafíos en proyectos de aprendizaje automático
El aprendizaje automático nos brinda la capacidad de tomar decisiones más informadas y basadas en datos que son más rápidas que los enfoques convencionales. Sin embargo, como ocurre con cualquier otro método, el proceso de aprendizaje automático presenta su propio conjunto de desafíos.
### Desafíos con los datos
#### Datos de entrenamiento inadecuados
Tener datos suficientes para entrenar en un sistema es extremadamente crucial para el éxito en cualquier Proyecto de machine learning. Muchos de los sistemas de Machine Learning fallan estrepitosamente debido a la falta de datos.
Tener datos inadecuados significa que el sistema no puede comprender las tendencias adecuadamente, y esto compromete enormemente la eficiencia de ese sistema. Ésta es la razón por la que la recopilación de datos se ha convertido en una parte integral del proceso de aprendizaje automático. La cantidad de datos necesarios depende tanto de la complejidad del problema como del algoritmo seleccionado.
#### Datos de información no representativos
Para generalizar el modelo, es de vital importancia que los datos de entrenamiento sean una representación precisa de la población bajo consideración. Cada vez que se deriva un nuevo subconjunto de la población, es crucial que el subconjunto represente la imagen completa de ese conjunto de datos.
Si la muestra es demasiado pequeña, obtendremos ruido de muestreo, que se clasifica como datos no representativos. Por otra parte, incluso muestras grandes pueden resultar no representativas si el método de muestreo es defectuoso.
Esto se conoce como sesgo de muestreo. Es importante señalar que si reducimos el sesgo de muestreo, la varianza aumenta. En cuyo caso, el modelo no puede generalizar los datos invisibles; mientras que, si se reduce la varianza, se potencia el sesgo. Este fenómeno, también conocido como equilibrio entre sesgo y varianza, es el proceso de encontrar el equilibrio adecuado entre ambos parámetros.

### Calidad de los datos
La mala calidad de los datos es un gran desafío en los sistemas de aprendizaje automático. Estos sistemas requieren datos de alta calidad para evitar una situación en la que puedan fallar en cualquiera de las fases de capacitación y prueba. El aprendizaje automático es una técnica que utiliza muchos datos. Es por eso que la calidad de los datos utilizados en cualquier sistema de Machine Learning tiene un efecto enorme en su desarrollo. Debido a esto, pequeños errores en los datos de entrenamiento pueden provocar errores a gran escala en la salida del sistema. Los problemas cada vez más complejos requieren no sólo cantidades masivas de datos, sino también datos diversos. Debe ser completo y de buena calidad.

### Características apropiadas
Las características que se utilizan para entrenar el sistema afectan la eficiencia con la que logramos el resultado.
Las características irrelevantes pueden socavar cualquier sistema y también pueden aumentar el costo del entrenamiento.
La selección de características debe ser parte de la etapa más básica al diseñar un proyecto de aprendizaje automático.
Esto se vuelve aún más importante cuando uno tiene varias características. No es necesario utilizar todas las características disponibles mientras se desarrolla el algoritmo. Tenemos que optimizar nuestro algoritmo incorporando solo ciertas características que se consideran más importantes e independientes. A veces, menos es mejor (es decir, podemos usar la reducción de dimensionalidad manteniendo la variación), como en el caso de PCA, donde podemos usar los componentes principales como características. También podemos utilizar varias metodologías y técnicas para seleccionar un subconjunto del espacio de características para ayudar a que nuestros modelos funcionen mejor, porque más características generan más ruido.

## Sobreajuste y subajuste de los datos de entrenamiento

El sobreajuste se refiere a un sistema que modela demasiado bien los datos de entrenamiento. Esto sucede cuando un modelo aprende detalles innecesarios y gana ruido en los datos de entrenamiento. Esto afecta negativamente el rendimiento del sistema en los datos de prueba.

El subajuste ocurre cuando un sistema no es capaz de capturar con precisión las relaciones entre las características y las variables de salida. Esto afecta la capacidad del algoritmo para descifrar la tendencia subyacente de los datos.

**El subajuste es un fuerte indicador de que el algoritmo no es adecuado para el conjunto de datos considerado.**

El subajuste es común en los casos en que hay datos limitados disponibles para entrenar el sistema. Menos puntos de datos dan como resultado un entrenamiento no preciso del modelo que luego no funciona de manera eficiente en el conjunto de datos de prueba. Aumentar la cantidad de datos podría ser una forma de garantizar que el algoritmo tenga suficiente información para detectar tendencias y patrones generales.

# Regresión lineal

La regresión lineal es un algoritmo de aprendizaje supervisado que se utiliza para la predicción de variables continuas, como ventas, edad, precio del producto, salario y predicciones climáticas, etc. La regresión lineal está representada por una ecuación lineal en la que una variable es independiente (x ), y la segunda es la variable dependiente o la producción (y) (Smith et al., 1988).
Si los valores de entrada y salida están relacionados linealmente, entonces se puede dibujar una línea con los puntos de datos, como se muestra en la Figura.

![picture](https://drive.google.com/uc?id=1lWm-6YhIb8hE69LYvkd16ZFTT61rmAqL)

Esto describirá la relación entre las variables. En un modelo de regresión lineal simple con una sola x y una sola y, la ecuación sería la siguiente:

$y=mx+b$

Esto ilustra la ecuación de la recta, donde “x” es la variable independiente; “y” es la variable dependiente; “m” es la pendiente; “b” es la intersección con el eje y. La tarea de la regresión lineal es encontrar los valores óptimos de "m" y "b" basándose en los puntos de datos "x" y "y" de modo que la distancia promedio entre los puntos y la línea sea mínima. Para una línea, la distancia entre los puntos y las líneas se llama errores y cada punto tiene su propio error. Cuando estos errores se elevan al cuadrado y se suman, el valor así obtenido se denomina error cuadrático y la media de esta suma se denomina error cuadrático medio. Este error cuadrático medio es el error acumulativo entre todos los puntos de datos y la línea.

![picture](https://drive.google.com/uc?id=1ZrwnYzx-cKF2byAhKeeos3KpEoKi0XqF)

Como podemos observar en la Figura "a", puede haber muchos valores para “m” y “b”, la regresión encuentra la línea óptima de forma iterativa ajustando varias líneas y luego seleccionando la línea que tiene el menor error cuadrático medio (Figura b). Se puede seguir el mismo procedimiento para un problema donde hay más de dos variables; en cuyo caso, el problema se denomina regresión lineal multivariante.


Cuando la entrada tiene más de dos variables o tiene dimensiones superiores, entonces la recta de regresión es un plano para tres dimensiones o un hiperplano para más de tres dimensiones, donde la ecuación será:
y = m1x1 + m1x1 + m2x2 + m3x3 + m4x4 + + mnxn + b

En esta ecuación, y es la variable dependiente, y x1, x2, x3, x4,…, xn son variables o características independientes. m1, m2, m3, m4,…, mn son los coeficientes respectivos de las características, y estos presentan una idea sobre la proporción de cada característica para el cálculo de “y”. Por último, "b" es un término de sesgo.

A continuación, separamos las características y las etiquetas. Aquì las etiquetas son valores continuos ya que es un problema de regresión

```python
# contains all of the columns except for the last one.
X=dataset.iloc[:,:-1]
# is the independent variable
y = dataset.iloc[:,-1]
```

## Train-Test split


```python
from sklearn.model_selection import train_test_split
# The “train_test_split()” module has the features, labels, and splits ratio, where we have passed X, y, and 0.2, respectively.
# “random_state”is used, because the module “train_test_split” assigns data points to the training set and test set randomly.
# Passing the “random_state” argument will keep the randomness constant, and we will acquire the same split every time with a random state value.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
random_state=110)
```

Este módulo producirá cuatro variables (es decir, "X_train"), que contiene un conjunto de características que se utilizarán para entrenar el modelo con "y_train". Este, a su vez, contiene sus respectivas etiquetas, y "X_test" e "y_test" son parte de los datos de prueba. El modelo se evaluará utilizando estos.

## Training model


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Primero importamos la clase "LinearRegression" desde la biblioteca de sklearn y las instanciamos como "regressor".

Luego pasamos nuestras funciones de entrenamiento y las respectivas etiquetas al método "fit()".

El método "fit()" entrenará el modelo utilizando los datos proporcionados.

## Model evaluation

".predict()" predice valores en función de las caracteríticas proporcionadas.

Utilizamos este método para pronosticar valores basados en datos "X-test" y compararlos con etiquetas "y_test".


```python
import matplotlib.pyplot as plt
predictions=regressor.predict(X_test)
plt.scatter(y_test, predictions,color='k')
plt.xlabel('Test Data')
plt.ylabel('Predicted Data')
plt.title('Test Data Vs. Predicted Data')
plt.show()
```

Trazamos los valores predichos y los valores originales utilizando el método de dispersión de Matplotlib.

Para un modelo que tiene error cero, ontendremos una línea diagonal. En este caso los valores se distribuyen alrededor de la diagonal pero presentan una tendencia hacia diagonal.

La evaluación visual de este modelo no proporciona información exacta sobre los errores del modelo.

En este caso utilizamos medidas cuantitativas. Cuantitativamente, los errores son las diferencias entre los valores predichos y los valores originales.

Una medida estándar de los errores es la media o promedio de todas esas discrepancias.

Al evaluar modelos que predicen valores continuos, se utilizan tres tipos de valores cuantitativos:
- Mean Absolute Error (MAE), error medio absoluto.
  - El MAE mide la magnitud promedio de los errores en un conjunto de predicciones, sin considerar su dirección.
  - MAE de aproximadamente 39.22 significa que, en promedio, las predicciones de tu modelo están desviadas de los valores reales por alrededor de 39.22 unidades. Un MAE más bajo indica una mejor precisión del modelo.
- Mean Squared Error (MSE) (Error Cuadrático Medio):
  - El MSE mide la magnitud promedio de los errores al cuadrado. Esto tiende a penalizar los errores más grandes más que los errores más pequeños debido al cuadrado.
  - Un MSE de aproximadamente 2228.09 significa que, en promedio, los errores al cuadrado en tus predicciones suman aproximadamente 2228.09. Al igual que con el MAE, un MSE más bajo indica un mejor rendimiento del modelo.
- Root Mean Squared Error (RMSE) (Raíz del Error Cuadrático Medio):
  - El RMSE es simplemente la raíz cuadrada del MSE. Esto da una medida en la misma escala que la variable de interés original, lo que facilita la interpretación.
  - Un RMSE de aproximadamente 47.20 significa que, en promedio, las predicciones de tu modelo están desviadas de los valores reales por alrededor de 47.20 unidades. Al igual que con el MAE y el MSE, un RMSE más bajo indica un mejor rendimiento del modelo.




```python
from sklearn import metrics
print('Mean Absolute Error:',
metrics.mean_absolute_error(y_test,predictions))
print('Mean Squared Error:',
metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:',
np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

En el caso de regresiones lineales multivariables, como la que tenemos, podemos identificar la intercepción y los coeficientes de cada característica. Estos coeficientes pueden darnos una idea sobre la importancia de varias características.

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

La clase "LogisticRegression" es presente en la libreria "linear_model" del paquete sklearn.

### Evaluation of Model

Para evaluar un modelo, comparamos los valores recién generados con los originales. En el caso de una tarea de clasificación, estos valores se comparan entre sí en una matriz de confusión.

Una matriz de confusión esta constituida por los valores de verdaderos positivos, verdaderos negativos, falsos positivos y falsos negativos.

### Data Scaling
El escalado de datos hace que el rango de variables se vuelva uniforme. En este ejemplo, tenemos 12 funciones. Estudiemos su propagación utilizando la función “.describe()” de Pandas.

```python
dataset.describe().T
```

Para producir los datos en rangos similares, ejecutamos el escalado. Sklearn tiene varias clases de este tipo en su biblioteca de "preprocessing".

Primero, observaremos la clase “StandardScaler”, que resta todas las observaciones con sus medias y luego las divide con la desviación estándar

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_norm = sc.fit_transform(X)
X_scaled = pd.DataFrame(X_norm,columns=dataset.iloc[:,0:-1].columns)
X_scaled.describe().T
```

podemos ver que los valores medio, estándar, mínimo y máximo para valores continuos han cambiado. Ahora, las observaciones se escalan y se transforman en una distribución uniforme, que es la preferida por la mayoría de los algoritmos de aprendizaje automático. Podemos producir los datos para cada característica en un rango idéntico (es decir, entre 0 y 1) usando la clase "MinMaxScaler".

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_minmax = sc.fit_transform(X)
X_minmax = pd.DataFrame(X_minmax,columns=dataset.iloc[:,0:-1].columns)
X_minmax.describe().T
```

podemos observar que los valores están en rango idéntico si nos referimos a las columnas min y max. El valor mínimo es "0" y el valor máximo es "1" para todas las funciones.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size =
0.2, random_state = 0)
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

Una vez entrenado el modelo con datos escalados, podemos evaluarlo una vez más para confirmar si hemos mejorado el modelo:

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

Al escalar los datos obtuvimos una mejora del 2%, con respecto a los datos no escalados. Esto muestra que la regresión logística no es demasiado sensible al escalado de datos, a diferencia de ciertos algoritmos, como las redes neuronales artificiales, que son muy sensibles al escalado de datos.
