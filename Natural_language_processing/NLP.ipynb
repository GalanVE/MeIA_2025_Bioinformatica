## TF-IDF

Frecuencia de término - Frecuencia de documento inversa

- Es una medida que permite evaluar la importancia de una palabra en un documento en relación con una colección de documentos.

- A diferencia de la frecuencia simple TF-IDF equilibra palabras comunes y raras para resaltar los términos más significativos

- TF Mide la frecuencia con la que aparece una palabra en un documento.

- IDF Reduce la importancia de las palabras comunes en varios documentos y aumenta la de las palabras poco comunes. Si un término aparece en menos documentos, es más probable que sea significativo y especifico.

El IDF se calcula con la siguiente fórmula:

$IDF_{(word)}=log(D/I_{(word)})$


donde: $IDF_{(word)}$ es la importancia de una palabra en el documento; $D$ es el número total de instancias del corpus; $I_{(word)}$ es el número de instancias de texto de la palabra en particular.

- En pocas palabras, $IDF$ es el inverso del número de apariciones de una palabra en un corpus.

### Procesamiento de textos

Para el procesamiento de texto, necesitaremos una biblioteca adicional llamada “nltk” o kit de herramientas de lenguaje natural. Esta biblioteca contiene varios métodos para tratar datos de texto. En esta parte, utilizamos la biblioteca "nltk" para eliminar palabras vacías como "the", "I", "we", etc., que se consideran ruido porque tienen una alta frecuencia en cualquier dato de texto pero tienen una contribución mínima a la información sobre un texto. La biblioteca "nltk" se puede instalar utilizando el instalador "pip" o "conda".

```python
import nltk
nltk.download('stopwords')
```

Una vez instalada la biblioteca, descargaremos la lista de palabras vacías utilizando el método de descarga. El "nltk" también contiene varios corpus como conjuntos de datos de ejemplo para practicar tareas de PNL. Comprobemos el contenido de las “stopwords”.

```python
from nltk.corpus import stopwords
stopwords.words('english')[0:15]
```

En el resultado, se nos presentan las 15 palabras vacías principales de la lista. A continuación, escribiremos una función para eliminar estas palabras vacías del texto de los resúmenes.

```python
def text_processing(text):
  word_list = text.lower().split(" ")
  processed_words = []
  for word in word_list:
    if word not in stopwords.words('english'):
      processed_words.append(word)
  return " ".join(processed_words)
```

Esta es una función simple en la que tomamos el texto como entrada y luego lo dividimos en una lista de palabras.
Estas palabras se examinan para detectar la presencia de palabras en el corpus de "stopword". Las palabras que no se encuentran en el corpus se retienen y se unen para formar un texto.

```python
text = 'Hi we are discussing NLP in this notebook'
text_processing(text)
```

En el resultado, podemos observar que la función ha eliminado "we", "are", "in" y "this" del texto original, y el nuevo texto a analizar ahora contiene palabras clave más informativas. En el siguiente paso, aplicaremos esta función a la columna "abstract" del conjunto de datos.

```python
X = dataset['Abstract'].apply(text_processing)
dataset['Abstract'] = X
```

Como el texto de los resúmenes ahora excluye las palabras vacías, separaremos las variables dependientes e independientes.

```python
X = dataset['Abstract']
y = dataset['Label']
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y[:5]
```

Para convertir nuestras etiquetas en representaciones numéricas para cada clase, utilizamos la clase "LabelEncoder" de sklearn. Como podemos ver, los primeros "cinco" elementos del resultado tienen números enteros para las clases.

## Bolsa de palabras de tokenización de texto
En esta sección, implementaremos la vectorización de texto. Convertiremos las oraciones al modelo de bolsa de palabras. Sklearn tiene ciertas bibliotecas de extracción de funciones importantes para datos de texto. La clase "CountVectorizer" convertirá los resúmenes al modelo de bolsa de palabras, donde cada columna representa las palabras y las entradas de las columnas representan el recuento de palabras de las palabras mencionadas anteriormente.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(X).toarray()
X.shape
```

Luego de transformar “X”, podemos observar que la matriz de datos es del tamaño 5,422 × 25,953, lo que implica que el corpus tiene 25,953 palabras en su vocabulario, y tenemos 5,422 instancias de texto. “25.953” es una dimensión significativamente grande. Apliquemos PCA para reducir la dimensión a "2" y visualizar las instancias de texto en un espacio 2D.

```python
from sklearn.decomposition import PCA
sklearn_pca = PCA(n_components=2)
PCs = sklearn_pca.fit_transform(X)
PCA_x = pd.DataFrame(PCs,columns=['PC1','PC2'])
PCA_x = pd.concat([PCA_x,dataset['Label']],axis=1)
fig, axes = plt.subplots(figsize=(10,8))
sns.set_style('whitegrid')
sns.scatterplot(x='PC1',y='PC2',data = PCA_x,hue='Label')
```

Podemos observar que el resumen de la clase de enfermedad viral está más disperso y que no se puede recuperar información suficiente de este gráfico PCA. Introduzcamos los datos vectorizados en el algoritmo de regresión logística y estudiemos cómo funciona con estos datos.

## Splitting the Dataset into the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
y,test_size=0.3,random_state=101)
```

## Building Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Model Evaluation

```python
predictions = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

La regresión logística pudo priorizar las características de las clases respectivas y arrojó una tasa de precisión del 88%.

## Implementación del TF-IDF
Ahora implementaremos TF-IDF utilizando la clase de sklearn “TFidFTransformer”. Recopila los datos del recuento de palabras (es decir, los datos transformados utilizando el objeto "CountVectorizer") y aplica TF-IDF a cada instancia de texto.

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X = tfidf_transformer.fit_transform(X)
print(X.shape)
X = X.toarray()
```

Podemos recuperar el IDF (es decir, frecuencias inversas de documentos) de palabras del diccionario “cv.vocabulary_”. En este diccionario, las palabras son las claves y los valores representan las puntuaciones de la IDF.

```python
print(tfidf_transformer.idf_[cv.vocabulary_['lungs']])
```

El resultado muestra la puntuación IDF para la palabra "lungs". Nuevamente, trazamos PCA transformando las características de TFIDF en dos dimensiones.

```python
from sklearn.decomposition import PCA
sklearn_pca = PCA(n_components=2)
PCs = sklearn_pca.fit_transform(X)
PCA_x = pd.DataFrame(PCs,columns=['PC1','PC2'])
PCA_x = pd.concat([PCA_x,dataset['Label']],axis=1)
fig, axes = plt.subplots(figsize=(10,8))
sns.set_style('whitegrid')
sns.scatterplot(x='PC1',y='PC2',data = PCA_x,hue='Label')
```

En este gráfico, podemos ver la similitud de instancias dentro de las clases, así como las diferencias entre las clases. A continuación, aplicaremos los datos transformados de TF-IDF para entrenar nuevamente la regresión logística.

## Splitting the Dataset into the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

## Building Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Model Evaluation

```python
predictions = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

Esta vez, obtuvimos una tasa de precisión inferior al problema anterior en un 2%. Esto podría atribuirse al hecho de que estamos contabilizando 5.422 instancias de texto (que es mucho menos que la cantidad de características o vocabulario en comparación con 25.953) para el cálculo de las puntuaciones TF-IDF. A medida que aumenta el corpus, el modelo TF-IDF funcionará mejor que las transformaciones CountVectorizer.

# Artificial Neural Networks in NLP

En esta sección, implementaremos redes neuronales para el problema de clasificación. La aplicación de redes neuronales en PNL también se conoce como PNL profunda. Como los textos vectorizados tienen muchas características, se prefieren las redes neuronales a otros algoritmos.

## Binarización de etiquetas

- Para aplicar el concepto de ANN a PNL, cambiaremos las etiquetas al formato de uno contra todos.


<img src="../Imagen/vectortext2.png" width="600" height="350"/>


```python
# Sklearn’s “LabelBinarizer” transforms the labels into binary classes.

y = dataset['Label']
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
y = lb.fit_transform(y)
y.shape
```

El atributo class_ del objeto "LabelBinarizer" almacena la matriz de todas las clases originales para que podamos recuperarla en el futuro durante la predicción.

```python
lb.classes_
```

## Splitting the Dataset into the Training Set and the Test Set

Dividamos las características de transformación TF-IDF y las etiquetas binarizadas en conjuntos de entrenamiento y prueba.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,
random_state=101)
```

## Building Model

Construiremos el modelo ANN utilizando la biblioteca Keras de TensorFlow.

```python
!pip install tensorflow
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
model = Sequential()
#
model.add(Dense(units=32,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=4,activation='softmax'))
# For a binary classification problem
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=
['accuracy'])
```

El modelo tiene tres capas. La primera capa contiene 32 neuronas que están densamente conectadas a la segunda capa que tiene 16 neuronas, y la última capa tiene cuatro neuronas porque tenemos la tarea de predecir cuatro clases.
La función de activación de la última capa es "softmax" y hemos aplicado el optimizador "adam". Hay abandonos de 0.5 entre las capas. El modelo se entrenará utilizando el objeto de devolución de llamada EarlyStopping.

## Training Model

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1,
patience=1)
model.fit(x=X_train,
y=y_train,
epochs=400,
batch_size=128,
validation_data=(X_test, y_test), verbose=1,
callbacks=[early_stop]
)
```

## Model Evaluation

En este momento, evaluaremos el modelo con datos de prueba.

```python
predictions = model.predict(X_test)
predictions = lb.inverse_transform(predictions)
y_test = lb.inverse_transform(y_test)
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

El modelo logró una precisión del 91 % en el conjunto de prueba, lo que es un 35 % mejor que el mejor de los dos modelos anteriores.
